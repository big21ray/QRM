```r
residuals_garch <- function(log_losses) {
    
    

    log_NA <- na.omit(log_losses)
    spec <- ugarchspec(variance.model=list(model="sGARCH"),mean.model=list(armaOrder=c(0,0)))
    
    fit <- ugarchfit(spec, data = log_NA)
    res <- (residuals(fit, standardize = TRUE))

    return(res)
}
```

```r
res.AMZN  = residuals_garch(df_AMZN$AMZNLogReturns)
res.NKE  = residuals_garch(df_NKE$NKELogReturns)
res.UAA  = residuals_garch(df_UAA$UAALogReturns)
res.BIDU  = residuals_garch(df_BIDU$BIDULogReturns)
```

```r
garch <- function(log_losses) {
    
    

    log_NA <- na.omit(log_losses)
    spec <- ugarchspec(variance.model=list(model="sGARCH"),mean.model=list(armaOrder=c(0,0)))
    
    fit <- ugarchfit(spec, data = log_NA)
    res <- (residuals(fit, standardize = TRUE))

    return(fit)
}
```

```r
mu.AMZN <- as.numeric(fitted(fit.AMZN)[1])
sigma.AMZN <- ugarchforecast(fit.AMZN, n.ahead = 1)@forecast$sigmaFor[1] 

mu.NKE <- as.numeric(fitted(fit.NKE)[1])
sigma.NKE <- ugarchforecast(fit.NKE, n.ahead = 1)@forecast$sigmaFor[1] 

mu.BIDU <- as.numeric(fitted(fit.BIDU)[1])
sigma.BIDU <- ugarchforecast(fit.BIDU, n.ahead = 1)@forecast$sigmaFor[1] 

mu.UAA <- as.numeric(fitted(fit.UAA)[1])
sigma.UAA <- ugarchforecast(fit.UAA, n.ahead = 1)@forecast$sigmaFor[1] 
res <- cbind(res_AMZN, res_BIDU, res_NKE, res_UAA)

```

```r
mu__ <- cbind(mu.AMZN,mu.BIDU,mu.NKE,mu.UAA)
sigma__ <- cbind(sigma.AMZN,sigma.BIDU,sigma.NKE,sigma.UAA)
```

```r
sim_returns <- matrix(nrow = N, ncol = 4)

for (i in 1:4){
    for (j in 1:N){
        sim_returns[j,i] = mu__[i] + sigma__[i]*quantile(res[,i],sim[j,i])
        
    }
}
```

```r
sim_return2 <- matrix(nrow = N, ncol = 4)

for (i in 1:4){
    for (j in 1:N){
        percentile <- ecdf(res[,i])
        q <- percentile(sim[j,i])
        sim_return2[j,i] = mu__[i] + sigma__[i]*q
        
    }
}
```

```r

L2 <- matrix(nrow = N, ncol = 1)

for (j in 1:N){
    p <- 0
    for (i in 1:4){
        p <- p + sim_return2[j,i] 
    }
    p <- p/4
    L2[j] <- -p
    
}
L2
```
